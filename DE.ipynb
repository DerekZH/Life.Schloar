#### Getting general topics for LDA results

```python
def get_general_topics(df_row, threshold = 0.2):
    general_topic = ''
    for key, value in df_row.to_dict().items():
        if key not in ['course_name', 'color', 'tsne_1', 'tsne_2', 'provider']:
            if value > threshold:
                if len(general_topic) == 0:
                    general_topic = key
                else:
                    general_topic = general_topic + ', ' + key
    return general_topic
```

```python
LDA_results = pd.read_csv('Udemy_Coursera_4000_LDA_results.csv')
```

```python
LDA_results.drop('general_topics', axis=1, inplace=True)
```

```python
LDA_results['general_topics'] = LDA_results.apply(get_general_topics, axis=1)
```

```python
collector = []
for row in lda.iloc[:,1:27].iterrows():
    dic = {}
    for key, value in row[1].to_dict().items():
        if value > 0.2:
            dic[key] = value
    for item in combinations(dic.keys(), 2):
        collector.append(item)

    
```

```python
counter = {}

for item in collector:
    try:
        counter[item] += 1
    except:
        counter[item] = 1
```

```python
import operator
sorted_counter = sorted(counter.items(), key=operator.itemgetter(1), reverse=True)
```

```python
sorted_counter_excluding_online_ed = []
for item in sorted_counter:
    if 'online_ed_related' not in item[0]:
        sorted_counter_excluding_online_ed.append(item)
```

```python
keys = ()
for item in sorted_counter_excluding_online_ed:
    keys = keys + item[0]
keys = set(keys)
keys = sorted(keys)
```

```python
topic_num_dic = dict(zip(keys, range(len(keys))))
```

```python
import numpy as np
from ast import literal_eval
```

```python
cooccurance_matrix = np.zeros((len(keys), len(keys)))
```

```python
sorted_counter_excluding_online_ed_and_dedup = {}
for item in sorted_counter_excluding_online_ed:
    try:
        sorted_counter_excluding_online_ed_and_dedup[str(sorted(item[0]))] += item[1]
    except:
        sorted_counter_excluding_online_ed_and_dedup[str(sorted(item[0]))] = item[1]
```

```python
for key, value in sorted_counter_excluding_online_ed_and_dedup.items():
    i, j = literal_eval(key)
    i, j = topic_num_dic[i], topic_num_dic[j]
    cooccurance_matrix[i,j] = value
    cooccurance_matrix[j,i] = value
    
```

```python
cooccurance_matrix = pd.DataFrame(cooccurance_matrix, columns=keys, index=keys)
```

```python
sns.heatmap(cooccurance_matrix, linewidths=0.5)
```

```python
topic_sum = {}

for col in lda.iloc[:,1:27]:
    topic_sum[col] = lda[col].sum()
```

```python
topic_sum
```

```python
cooccurance_matrix_normalized = np.zeros((len(keys), len(keys)))
```

```python
# normalizing co-occurance:
for key, value in sorted_counter_excluding_online_ed_and_dedup.items():
    topic_i, topic_j = literal_eval(key)
    i, j = topic_num_dic[topic_i], topic_num_dic[topic_j]
    cooccurance_matrix_normalized[i,j] = value / topic_sum[topic_i] / topic_sum[topic_j]
    cooccurance_matrix_normalized[j,i] = value / topic_sum[topic_i] / topic_sum[topic_j]
```

```python
cooccurance_matrix_normalized = pd.DataFrame(cooccurance_matrix_normalized, columns=keys, index=keys)
```

```python
sns.heatmap(cooccurance_matrix_normalized, 0,0.0040, linewidths=0.5)
```

```python
# topic = keys[randint(0,26)]
related_topics_dic = {}
similarity_cut_off = 0.0008

for i in range(cooccurance_matrix_normalized.shape[0]):
    topic = keys[i]
    related_topics_dic[topic] = cooccurance_matrix_normalized[topic].loc[cooccurance_matrix_normalized[topic] > similarity_cut_off].index.values.tolist()
```

```python
with open('LDA_related_topics.pickle', 'wb') as f:
    pickle.dump(related_topics_dic, f)
```
